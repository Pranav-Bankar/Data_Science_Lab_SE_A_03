```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers."
    elif n == 0:
        return 1
    else:
        res = 1
        for i in range(1, n + 1):
            res *= i
        return res

# Take input from the user
while True:
    try:
        user_input = input("Enter a non-negative integer to compute its factorial: ")
        number = int(user_input)
        if number < 0:
            print("Please enter a non-negative integer.")
        else:
            print(f"The factorial of {number} is {factorial(number)}")
            break
    except ValueError:
        print("Invalid input. Please enter an integer.")
```

```python
# Function to display the multiplication table
def display_multiplication_table(number):
    print(f"Multiplication Table for {number}:")
    for i in range(1, 11):  # Table up to 10
        print(f"{number} x {i} = {number * i}")

# Take input from the user
while True:
    try:
        user_input = input("Enter a number to display its multiplication table: ")
        num = int(user_input)
        display_multiplication_table(num)
        break
    except ValueError:
        print("Invalid input. Please enter an integer.")
```

```python
# Function to check if a number is odd or even
def check_odd_even(number):
    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"

# Take input from the user
while True:
    try:
        user_input = input("Enter an integer to check if it's odd or even: ")
        num = int(user_input)
        result = check_odd_even(num)
        print(f"The number {num} is {result}.")
        break
    except ValueError:
        print("Invalid input. Please enter an integer.")
```

```python
# Function to check if a number is a palindrome
def is_palindrome(number):
    # Convert number to string to easily reverse and compare
    num_str = str(number)
    return num_str == num_str[::-1]

# Take input from the user
while True:
    try:
        user_input = input("Enter an integer to check if it's a palindrome: ")
        num = int(user_input)

        if is_palindrome(num):
            print(f"The number {num} is a palindrome.")
        else:
            print(f"The number {num} is not a palindrome.")
        break
    except ValueError:
        print("Invalid input. Please enter an integer.")
```

# Task
Create a simple expert system that recommends an activity based on user input for weather and time, demonstrating a knowledge base, an inference engine, and interactive input/output.

## Define Knowledge Base

### Subtask:
Establish the rules and facts for a simple decision-making scenario (e.g., recommending an activity based on weather and time). This will involve creating data structures to store the knowledge.


**Reasoning**:
To establish the rules and facts for the decision-making scenario, I will define a Python dictionary to act as the knowledge base. This dictionary will map combinations of weather and time conditions to a recommended activity, as specified in the instructions.



```python
knowledge_base = {
    ('sunny', 'day'): 'Go for a walk',
    ('rainy', 'day'): 'Read a book indoors',
    ('cloudy', 'day'): 'Visit a museum',
    ('sunny', 'night'): 'Stargaze',
    ('rainy', 'night'): 'Watch a movie',
    ('cloudy', 'night'): 'Play board games'
}

print("Knowledge base created:")
for (weather, time), activity in knowledge_base.items():
    print(f"  If weather is '{weather}' and time is '{time}', recommend: '{activity}'")
```

## Implement Inference Engine

### Subtask:
Develop a basic inference engine using Python logic (e.g., if-elif-else statements) to process the defined rules and derive a decision based on user input.


**Reasoning**:
To implement the inference engine as requested, I will define a Python function `get_recommendation` that retrieves an activity from the `knowledge_base` based on `weather` and `time` inputs, and handles cases where no recommendation is found.



```python
def get_recommendation(weather, time):
    # Convert inputs to lowercase to match knowledge base keys consistently
    weather_lower = weather.lower()
    time_lower = time.lower()

    # Form a key tuple to look up in the knowledge base
    key = (weather_lower, time_lower)

    # Look up the recommendation in the knowledge base
    if key in knowledge_base:
        return knowledge_base[key]
    else:
        return "No specific recommendation available for these conditions. Try something else!"

print("Inference engine function 'get_recommendation' defined.")
```

**Reasoning**:
To complete the inference engine, I will add interactive input for weather and time, and then use the `get_recommendation` function to provide an activity recommendation to the user.



```python
print("\n--- Activity Recommender --- ")
while True:
    user_weather = input("Enter current weather (sunny, rainy, cloudy): ").lower()
    if user_weather not in ['sunny', 'rainy', 'cloudy']:
        print("Invalid weather. Please choose from 'sunny', 'rainy', 'cloudy'.")
        continue

    user_time = input("Enter time of day (day, night): ").lower()
    if user_time not in ['day', 'night']:
        print("Invalid time. Please choose from 'day', 'night'.")
        continue

    recommendation = get_recommendation(user_weather, user_time)
    print(f"Based on '{user_weather}' weather and '{user_time}' time, I recommend: {recommendation}")
    break

print("Activity recommendation process completed.")
```
